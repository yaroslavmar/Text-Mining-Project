## Get links of the speeches


```python
# Get links of speeches

url="https://www.ecb.europa.eu/press/key/speaker/pres/html/index.en.html"
html = urllib.urlopen(url).read()
soup = BeautifulSoup(html)

blocks=soup.select(".ecb-langSelector")
del blocks[0]

path=[blocks[x]('a')[0]['href'] for x in range(len(blocks))]

links=["https://www.ecb.europa.eu" + x for x in path]

```

```python
# Get Author, Title and date

# Author and title
top=soup.select(".doc-title")
author=[]
title=[]

for x in top:
        top_split=x.text.split(":")
        author.append(top_split[0])
        title.append(top_split[1])
        
author=[str(unicodedata.normalize('NFKD', x).encode('ascii','ignore')) for x in author] 
author=[x.strip('\n') for x in author]

title=[str(unicodedata.normalize('NFKD', x).encode('ascii','ignore')) for x in title]
title=[x.replace ( "\n", "" ) for x in title] 

x.replace ( "\n", "" )
# Date
date=soup.select("dt")
date=[x.text for x in date]
date=[str(unicodedata.normalize('NFKD', x).encode('ascii','ignore')) for x in date]

# Index
index=range(577,0,-1)
```

## Get speech and save as JSON file

```python

# Input: Url of speech
# Output 
def GetSpeech(speech_url):
    html=urllib.urlopen(speech_url).read()
    soup = BeautifulSoup(html)
    speech_text=soup.select("article p")
    speech=[x.text for x in speech_text]
    speech=[str(unicodedata.normalize('NFKD', x).encode('ascii','ignore')) for x in speech]
    speech=" ".join(speech)
    return speech



```
